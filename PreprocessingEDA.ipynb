```python
#required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as snb
```

Checking Missing values

```python
df.isnull().sum()
```

```python
#due to the lack of available data in tags and consumer disputed removing from df
df = df.drop(['Tags','Consumer disputed?'],axis=1) 
```

renaming the columns 

```python
col_mapping = {'Consumer complaint narrative':'Consumer_complaint','Company public response':'Company_response','ZIP code':'Zip_code',
               'Consumer consent provided?':'Consent provided','Company response to consumer':'Company_response',
               'Timely response?':'Timely_response'}
df.rename(columns=col_mapping, inplace=True)
df.columns
```

Cleaning the missing values 

```python
df_new = df.dropna(subset=['Sub-product', 'Issue', 'Sub-issue',
       'Consumer_complaint', 'Company_response', 'Company', 'State',
       'Zip_code', 'Consent provided', 'Submitted via', 'Date sent to company',
       'Company_response', 'Timely_response', 'Complaint ID'])
```

```python
df_new.isnull().sum()
```

```python
df['Date received'] = pd.to_datetime(df['Date received'], format='%m/%d/%Y',errors='coerce')
df['Date sent to company'] = pd.to_datetime(df['Date sent to company'], format='%m/%d/%Y',errors='coerce')
```

Data Visualization
