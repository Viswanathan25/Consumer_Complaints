**2. Given an unsorted array of integers, find the length of the longest continuous increasing subsequence (subarray).**

Example 1:

Input: [1,3,5,4,7]

Output: 3

Example 2:

Input: [2,2,2,2,2]

Output: 1

```python
def longest_sub_seq(seq):   # Defined the function longest_sub_seq to find longest continuous increasing subsequence

  count = 1          # Initialised count object with int value '1'

  for i in range(len(seq)-1):  # for loop is used to iterate values
    if seq[i]<seq[i+1]:    # Checking condition whether current is smaller than next , if it is add 1 to the "count"
      count+=1
    else:
      break
  return count


longest_sub_seq(list(map(int,input().split())))  # Getting list of values from users and passing  as arguments to the parameter
```

**3. Given a list of non negative integers, arrange them such that they form the largest number**.

 Example 1:

 Input: [10,2]

 Output: "210"

 Example 2:

 Input: [3,30,34,5,9]

 Output: "9534330"

```python
from functools import cmp_to_key

def largestNumber(nums):
    # Convert the numbers to strings for easier comparison
    nums_str = list(map(str, nums))

    # Custom comparator to decide which combination is greater
    def compare(x, y):
        # Compare combined numbers in both possible orders
        if x + y > y + x:
            return -1
        elif x + y < y + x:
            return 1
        else:
            return 0

    # Sort the numbers with the custom comparator
    nums_str.sort(key=cmp_to_key(compare))

    # Join the sorted strings
    largest_num = ''.join(nums_str)

    # Handle the case where we have leading zeros (e.g., [0, 0])
    return '0' if largest_num[0] == '0' else largest_num


print(largestNumber(list(map(int,input().split(',')))))
```
